### PROS & CONS

### Pros:

Improves dataset completeness: General imputation helps to fill in missing values in a dataset, thereby improving the completeness of the data. This can lead to more accurate and reliable analyses.

Preserves sample size: By filling in missing values, general imputation can preserve the sample size of a dataset, which can be important for statistical power and avoiding biased estimates.

Simplifies analysis: General imputation can simplify the analysis of a dataset by providing a complete dataset for analysis, rather than having to exclude observations with missing values.

### Cons:

May introduce bias: General imputation may introduce bias into the dataset if the imputation method used is not appropriate for the missing values or the underlying distribution of the data.

Ignores uncertainty: General imputation assumes that the imputed values are accurate representations of the true values of the missing data, which may not be the case. This can introduce uncertainty into the analysis.

Underestimates variability: General imputation can underestimate the variability in the dataset, as it may not accurately capture the true variability of the missing data.

Can be time-consuming: General imputation can be a time-consuming process, particularly if there are a large number of missing values or if complex imputation methods are used.

Overall, general imputation can be a useful tool for dealing with missing data in a dataset, but it is important to carefully consider the advantages and disadvantages of different imputation methods and to be cautious when interpreting imputed values.

### Conclusion 

In conclusion, general imputation is a statistical technique used to fill in missing values in a dataset. It can be a valuable tool for improving dataset completeness and preserving sample size, and can simplify the analysis of a dataset by providing a complete dataset for analysis. However, general imputation can introduce bias, ignore uncertainty, and underestimate variability, and may be time-consuming to implement. As such, it is important to carefully consider the advantages and disadvantages of different imputation methods and to use caution when interpreting imputed values. Ultimately, general imputation can be a useful approach to dealing with missing data in a dataset, but should be used judiciously and in conjunction with other methods to ensure accurate and reliable analyses.

# 2. Imputation with Linear Regression

#### Predicting missing values of Data by Linear Regression Model

Imputation with linear regression is a common method used to fill in missing values in a dataset. It involves using a linear regression model to predict the missing values based on the values of other variables in the dataset. The linear regression model is trained on the non-missing values in the dataset, and then used to predict the missing values.

The basic steps involved in imputation with linear regression are as follows:

Identify the variables with missing values in the dataset.
Split the dataset into two sets: a set with complete data (i.e., no missing values) and a set with missing data.
Create a linear regression model using the complete data set as the training set.
Use the linear regression model to predict the missing values in the dataset with missing data.
Merge the two sets back together to create a complete dataset with no missing values.
Imputation with linear regression can be a powerful tool for dealing with missing values in a dataset, as it can provide accurate predictions of missing values based on other variables in the dataset. However, it is important to keep in mind that the accuracy of the imputed values depends on the quality of the linear regression model and the characteristics of the dataset. Additionally, imputation with linear regression can be computationally intensive, particularly for large datasets with many missing values.

## 3. k-Neareast Neighbor (kNN) Imputation

The k-nearest neighbors algorithm, also known as KNN or k-NN, is a non-parametric, supervised learning classifier, which uses proximity to make classifications or predictions about the grouping of an individual data point.

1. K-Nearest Neighbour is one of the simplest Machine Learning algorithms based on Supervised Learning technique.

2. K-NN algorithm assumes the similarity between the new case/data and available cases and put the new case into the category that is most similar to the available categories.

3. K-NN algorithm stores all the available data and classifies a new data point based on the similarity. This means when new data appears then it can be easily classified into a well suite category by using K- NN algorithm.

4. K-NN algorithm can be used for Regression as well as for Classification but mostly it is used for the Classification problems.

5. K-NN is a non-parametric algorithm, which means it does not make any assumption on underlying data.
It is also called a lazy learner algorithm because it does not learn from the training set immediately instead it stores the dataset and at the time of classification, it performs an action on the dataset.

6. KNN algorithm at the training phase just stores the dataset and when it gets new data, then it classifies that data into a category that is much similar to the new data.

The K-NN working can be explained on the basis of the below algorithm:

Step-1: Select the number K of the neighbors

Step-2: Calculate the Euclidean distance of K number of neighbors

Step-3: Take the K nearest neighbors as per the calculated Euclidean distance.

Step-4: Among these k neighbors, count the number of the data points in each category.

Step-5: Assign the new data points to that category for which the number of the neighbor is maximum.

Step-6: Our model is ready.

```python
dataset_knn = {"Height":[167,182,176,173,172,174,169,173,170,170],
               "Weight":[51,62,69,64,65,56,58,57,55,57],
               "Class":["Underweight","Normal","Normal","Normal","Normal","Underweight", "Normal","Normal","Normal", "Unknown"],
                }
df_knn = pd.DataFrame(dataset_knn)
df_knn
```

KNN is a simple machine learning algorithm that can be used for classification and regression problems. In classification problems, the goal is to predict the class label of an unknown instance based on its proximity to the instances in the training data. In this case, the unknown instance is compared to the instances in the training data using a distance metric, such as the Euclidean distance.

The Euclidean distance is a measure of the straight-line distance between two points in a multi-dimensional space. In this code, the Euclidean distance between the height and weight of an unknown individual and the height and weight of each individual in the dataset is calculated and stored in a new column in the DataFrame. This information can then be used to classify the unknown individual based on the k-nearest neighbors.

```python
dataset_knn = {"Height":[167,182,176,173,172,174,169,173,170,170],
               "Weight":[51,62,69,64,65,56,58,57,55,57],
               "Class":["Underweight","Normal","Normal","Normal","Normal","Underweight", "Normal","Normal","Normal", "Unknown"],
                }
#Height and weight of unkown data
x2 = 170
y2 = 57
distances = []
for i in range(len(dataset_knn["Height"])):
    x1 = dataset_knn["Height"][i]
    y1 = dataset_knn["Weight"][i]
    d = ((x2 - x1)**2 + (y2 - y1)**2)**0.5
    distances.append(d)
dataset_knn["Euclidean distance"] = distances
df = pd.DataFrame(dataset_knn)
df
```

## 4.  Polynomial Regression Model

1. Polynomial Regression is a form of regression analysis in which the relationship between the independent variables and dependent variables are modeled in the nth degree polynomial.

2. Polynomial Regression models are usually fit with the method of least squares.The least square method minimizes the variance of the coefficients,under the Gauss Markov Theorem.

3. Polynomial Regression is a special case of Linear Regression where we fit the polynomial equation on the data with a curvilinear relationship between the dependent and independent variables.

#### 3. Types of Polynomial Regression
A quadratic equation is a general term for a second-degree polynomial equation. This degree, on the other hand, can go up to nth values. Polynomial regression can so be categorized as follows:

1. Linear – if degree as 1

2. Quadratic – if degree as 2

3. Cubic – if degree as 3 and goes on, on the basis of degree

#### 3. Assumption of Polynomial Regression
    

1. The behaviour of a dependent variable is explained by a linear, or curvilinear, additive relationship between the dependent variable and a set of k independent variables (xi, i=1 to k).
2. The relationship between the dependent variable and any independent variable is linear or curvilinear.
3. The independent variables do no depend on each other too.
4. The errors are independent, normally distributed with mean zero and a constant variance.

#### 4. Why do we need polynomial regression?

1. We make our model and find out that it performs very badly.

2. We observe between the actual value and the best fit line,which we predicted and it seems that the actual value has some kind of curve in the graph and our line is no where near to cutting the mean of the points.

3. This where polynomial Regression comes to the play, it predicts the best fit line that follows the pattern(curve) of the data.

4. Polynomial Regression does not require the relationship between the independent and dependent variables to be linear in the data set,This is also one of the main difference between the Linear and Polynomial Regression.

5. Polynomial Regression is generally used when the points in the data are not captured by the Linear Regression Model and the Linear Regression fails in describing the best result clearly.

6. Increase the degree in the model,it tends to increase the performance of the model.However,increasing the degrees of the model also increases the risk of over-fitting and under-fitting the data.
